```python
#Plot each well by year
#['NO 15/9-F-1 C', 'NO 15/9-F-11 H', 'NO 15/9-F-12 H','NO 15/9-F-14 H', 'NO 15/9-F-15 D']

def pltcolor(lst):
    cols=[]
    for l in lst:
        if l =='NO 15/9-F-1 C':
            cols.append('red')
        elif l =='NO 15/9-F-11 H':
            cols.append('green')
        elif l =='NO 15/9-F-12 H':
            cols.append('blue')
        elif l =='NO 15/9-F-14 H':
            cols.append('orange')
        elif l =='NO 15/9-F-15 D':
            cols.append('black')
        
    return cols
# Create the colors list using the function above
cols=pltcolor(data_well['WELL_BORE_CODE'])

data_well.plot(kind='scatter',x='DATEPRD',y='BORE_OIL_VOL',c=cols)
#plt.xticks(datetime('01-Jan-08'),datetime('31-12-16'),1)
plt.show()
```
