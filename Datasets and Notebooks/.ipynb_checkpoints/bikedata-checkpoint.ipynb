```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Reading data and adding month column
df1 = pd.read_csv('201901-citibike-tripdata.csv')
df1['Month']='Jan'
df1.head()
```

```python
#Reading data and adding month column
df2 = pd.read_csv('201902-citibike-tripdata.csv')
df2['Month']='Feb'
df2.head()
```

```python
#Reading data and adding month column
df3 = pd.read_csv('201903-citibike-tripdata.csv')
df3['Month']='Mar'
df3.head()
```

```python
#Calculating trip_min (trip duration) in minutes froms seconds
df1['trip_Min']=df1['tripduration']/60
df2['trip_Min']=df2['tripduration']/60
df3['trip_Min']=df3['tripduration']/60
```

```python
#Joining the three dataframes to get a combined dataframe for Jan-Mar 2019
df = pd.concat([df1, df2,df3], join="inner")
df.head()
```

```python
#Group by to calculate no of usertypes
df_group = df.groupby(['usertype'], sort=False).size().reset_index(name='Count')
df_group
```

```python
df.info()
```

```python
#Filtering dataframe for trip duration > 50 min and less than 4500 mins
df = df.loc[df['trip_Min'] > 50]
df = df.loc[df['trip_Min'] < 4500]
df.info()
```

```python
#Writing dataframe to csv
df.to_csv('bike_combined.csv')
```
