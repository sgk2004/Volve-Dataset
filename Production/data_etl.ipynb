## Exploratory Data Analysis (EDA) process

Pair Plots to see relationship between each columns

```python
import seaborn as sns

#make pairplots
sns_plot1=sns.pairplot(all_wells, vars = ["AVG_DOWNHOLE_PRESSURE","AVG_CHOKE_SIZE_P","AVG_WHP_P","BORE_OIL_VOL","BORE_GAS_VOL","BORE_WAT_VOL"], hue="WELL_BORE_CODE")
```

```python
sns_plot1.savefig("all_well.png")
```

```python
# Drop zero values in Average Downhole pressure
all_wells_drop_zero =all_wells[all_wells['AVG_DOWNHOLE_PRESSURE'] !=0]
```

```python
all_wells_drop_zero.info()
```

```python
all_wells_drop_zero.astype(bool).sum(axis=0)
```

```python
#make pairplots
sns_plot1a=sns.pairplot(all_wells_drop_zero, vars = ["AVG_DOWNHOLE_PRESSURE","AVG_CHOKE_SIZE_P","AVG_WHP_P","BORE_OIL_VOL","BORE_GAS_VOL","BORE_WAT_VOL"], hue="WELL_BORE_CODE")
```

```python
sns_plot1a.savefig("all_well_new.png")
```

```python
#Crest Wells
sns_plot2=sns.pairplot(discovery_wells, vars = ["AVG_DOWNHOLE_PRESSURE","AVG_CHOKE_SIZE_P","AVG_WHP_P","BORE_OIL_VOL","BORE_GAS_VOL","BORE_WAT_VOL"], hue="WELL_BORE_CODE")
```

```python
sns_plot2.savefig("discovery_well.png")
```
